```python

# packages
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display
from load_levels import load_levels
```

```python
# Load the hydrology data
file_path = '../data/river/Todmorden-Callis-Bridge-level-15min-Qualified.csv'
df = load_levels(file_path)

# Convert dateTime column to datetime format
df['dateTime'] = pd.to_datetime(df['dateTime'], errors='coerce')

# Add a 'season' column based on month
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df['season'] = df['dateTime'].dt.month.map(get_season)
df
```

```python
# simple plot

# Create a line plot of water level over time for multiple stations
fig = px.line(df, x='dateTime', y='value', color='station',
              labels={'dateTime': 'Date and Time', 'value': 'Water Level (m)', 'station': 'Monitoring Station'},
              title='Water Level Over Time by Monitoring Station')

# Show the plot
fig.show()
```
